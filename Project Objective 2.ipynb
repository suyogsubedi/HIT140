# Feature Engineering,Model Building, Linear Regression


Let's preprocess the data, perform feature engineering, and build linear regression models to predict motor and total UPDRS scores:



```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define features and target variables
X = data2.drop(columns=['subject#', 'motor_updrs', 'total_updrs'])
y_motor = data2['motor_updrs']
y_total = data2['total_updrs']

# Split the data into training and testing sets
X_train, X_test, y_motor_train, y_motor_test = train_test_split(X, y_motor, test_size=0.2, random_state=42)
X_train, X_test, y_total_train, y_total_test = train_test_split(X, y_total, test_size=0.2, random_state=42)

# Initialize linear regression models
motor_model = LinearRegression()
total_model = LinearRegression()

# Fit the models to the training data
motor_model.fit(X_train, y_motor_train)
total_model.fit(X_train, y_total_train)

# Predict on the test data
motor_predictions = motor_model.predict(X_test)
total_predictions = total_model.predict(X_test)

# Evaluate the models
motor_rmse = mean_squared_error(y_motor_test, motor_predictions, squared=False)
total_rmse = mean_squared_error(y_total_test, total_predictions, squared=False)
motor_r2 = r2_score(y_motor_test, motor_predictions)
total_r2 = r2_score(y_total_test, total_predictions)

print(f'Motor UPDRS RMSE: {motor_rmse}')
print(f'Total UPDRS RMSE: {total_rmse}')
print(f'Motor UPDRS R^2: {motor_r2}')
print(f'Total UPDRS R^2: {total_r2}')
```

# Analysis of the above output

In this analysis, two linear regression models were trained and evaluated to predict motor UPDRS and total UPDRS scores in subjects with Parkinson's disease. The code first splits the dataset into features (X) and target variables (y_motor for motor UPDRS and y_total for total UPDRS). Subsequently, it partitions the data into training and testing sets, allocating 80% of the data for training and 20% for testing, to assess model performance.

Two linear regression models, 'motor_model' and 'total_model,' were initialized and fitted to the training data. These models were then used to make predictions on the test data. The evaluation metrics used for both models include Root Mean Squared Error (RMSE) and R-squared (RÂ²).

The findings show that the motor UPDRS prediction model achieved an RMSE of approximately 7.48, indicating that, on average, the predicted motor UPDRS scores deviated from the actual scores by this amount. Additionally, the RÂ² value for the motor UPDRS model was approximately 0.12, suggesting that the model explains only a small portion of the variance in motor UPDRS scores.

Similarly, the total UPDRS prediction model produced an RMSE of around 9.66, signifying the average prediction error for total UPDRS scores. The RÂ² value for the total UPDRS model was approximately 0.16, implying a limited ability to explain the variability in total UPDRS scores.

These findings imply that the linear regression models, as currently configured, have limited predictive power for both motor UPDRS and total UPDRS scores, and more advanced modeling techniques or feature engineering may be required to improve their performance.

# Cross Validation

```python
from sklearn.model_selection import cross_val_score

# Cross-validation for motor UPDRS
motor_cv_scores = cross_val_score(motor_model, X, y_motor, cv=5, scoring='neg_mean_squared_error')
motor_cv_rmse = np.sqrt(-motor_cv_scores)

# Cross-validation for total UPDRS
total_cv_scores = cross_val_score(total_model, X, y_total, cv=5, scoring='neg_mean_squared_error')
total_cv_rmse = np.sqrt(-total_cv_scores)

print(f'Motor UPDRS Cross-Validation RMSE: {motor_cv_rmse.mean()}')
print(f'Total UPDRS Cross-Validation RMSE: {total_cv_rmse.mean()}')
```

# Cross Validation Analysis

In this analysis, cross-validation was performed to assess the predictive performance of the motor UPDRS and total UPDRS linear regression models. Cross-validation is a technique used to evaluate a model's ability to generalize to unseen data by splitting the dataset into multiple subsets, training the model on different combinations of these subsets, and evaluating its performance on each fold.

The code utilized the cross_val_score function from scikit-learn to conduct cross-validation with five folds (cv=5) for both the motor UPDRS and total UPDRS models. The scoring metric chosen for evaluation was negative mean squared error (neg_mean_squared_error), which measures the model's prediction error. By taking the negative of the mean squared error, we obtain a positive value that represents the quality of the model's predictions.

The findings indicate that the motor UPDRS model achieved an average cross-validation RMSE of approximately 8.97. This value represents the average prediction error across the five folds, suggesting that the model's predictions deviated from the actual motor UPDRS scores by this amount, on average.

Similarly, the total UPDRS model exhibited an average cross-validation RMSE of roughly 11.63, implying an average prediction error for total UPDRS scores across the five folds.

These findings further reinforce the earlier observations that the linear regression models, in their current state, have limited predictive power for both motor UPDRS and total UPDRS scores. Additional model refinement or feature engineering may be necessary to enhance their performance.

# Model Comparison

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

# Compare multiple regression models
models = {
    'Linear Regression': LinearRegression(),
    'Random Forest': RandomForestRegressor(random_state=42),
    'Support Vector Regressor': SVR()
}

param_grid = {
    'Random Forest': {'n_estimators': [50, 100, 200]},
    'Support Vector Regressor': {'C': [1, 10, 100], 'kernel': ['linear', 'rbf']}
}

best_models = {}

for model_name, model in models.items():
    if model_name in param_grid:
        grid_search = GridSearchCV(model, param_grid[model_name], cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
        grid_search.fit(X, y_motor)  # Change to y_total for total UPDRS
        best_model = grid_search.best_estimator_
        best_models[model_name] = best_model
    else:
        model.fit(X_train, y_motor_train)  # Change to y_total_train for total UPDRS
        best_models[model_name] = model

# Evaluate and compare best models
for model_name, best_model in best_models.items():
    predictions = best_model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_motor_test, predictions))  # Change to y_total_test for total UPDRS
    r2 = r2_score(y_motor_test, predictions)  # Change to y_total_test for total UPDRS
    print(f'{model_name} RMSE: {rmse}')
    print(f'{model_name} R^2: {r2}')
```
